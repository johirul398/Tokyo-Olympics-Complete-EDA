# 2.1 Load Data



```python
gender=pd.read_excel('../input/2021-olympics-in-tokyo/EntriesGender.xlsx')
```

# 2.2 Initial Idea About EntriesGender Data

```python
gender.info()
print('-'*100)
gender.head()
```

# 2.3 Stack Bar Plot  

```python
p=gender.sort_values(["Total"], ascending=False)
data = [p["Discipline"],p["Female"],p["Male"]]

headers = ["Discipline","Female", "Male"]

df3 = pd.concat(data, axis=1, keys=headers)

c=df3.head(15)
fig, ax = plt.subplots(figsize=(12,10))
c.plot(
  x = 'Discipline', 
  kind = 'barh', 
  stacked = True, 
  title = 'Stacked Bar Graph', 
  mark_right = True,ax=ax
)
total = c["Female"] + c["Male"]
rel = c[c.columns[1:]].div(total, 0)*100 
for n in rel:
    for i, (s, b, p) in enumerate(zip(c.iloc[:, 1:].cumsum(1)[n], 
                                         c[n], rel[n])):
        plt.text(s - b / 2, i, str(np.round(p, 1)) + '%', 
                 va = 'center', ha = 'center',fontsize = 10)
```

# 2.4 Count Male and Female Individually

```python
female = gender["Female"].sum()
male = gender["Male"].sum()
m=gender['Total'].sum()
d = {'Gender': ['Male', 'Female'],
        'Total' : [male,female]}
val = pd.DataFrame(d)
print(val)
print('*'*60)
sns.set()
val.set_index('Gender').T.plot(kind='bar', stacked=True)
```

# 2.5 Barplot

```python
sns.barplot(y =val['Total'],x=val['Gender'])
```

# 2.6 Pie chart

```python
female = gender["Female"].sum()
male = gender["Male"].sum()
m=gender['Total'].sum()
d =pd.DataFrame({'total': [male,female]},
     index=['Male', 'Female'])

def label_function(v):
    return f'{v:.0f}%'
     
d.plot.pie(y='total', figsize=(5, 5),colors=['#8B008B','#9932CC'],autopct=label_function,textprops={'fontsize': 20})
```

# 2.7 Group Bar Plot

```python
p=gender.sort_values(["Total"], ascending=False)
a=p.head(10)
d=a['Discipline']
m=a['Male']
f=a['Female']

x = np.arange(len(d)) 
width = 0.35 

fig, ax = plt.subplots(figsize=(15,5))
rects1 = ax.bar(x - width/2, m, width, label='Men')
rects2 = ax.bar(x + width/2, f, width, label='Female')

ax.set_ylabel('count')
ax.set_title('count by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(d)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

# 3.1 Load Data

```python
athlet=pd.read_excel('../input/2021-olympics-in-tokyo/Athletes.xlsx')
```

# 3.2 Initial Idea about Athletes Data

```python
athlet.info()
print('-'*100)
athlet.head(5)
```

# 3.3 Word Cloud

```python
from wordcloud import WordCloud, STOPWORDS
comment_words = ''
stopwords = set(STOPWORDS)
for val in athlet.Discipline:
    val = str(val)
    tokens = val.split()
     
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
 
wordcloud = WordCloud(width = 800, height = 500,
                background_color ='black',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
                     
plt.figure(figsize = (7, 10), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0) 
plt.show()
```

# 3.4 Top 20 Discipline

```python

count = athlet["Discipline"].value_counts().sort_values(ascending = False).reset_index()
f=count.head(20)
f.columns = ['count','athlet_type']
at_top_country=f
clrs = ['#FF4500' if (x < max(at_top_country['athlet_type'])) else 'black' for x in at_top_country['athlet_type'] ]
plt.figure(figsize = (10, 8))
sns.barplot(y =at_top_country['count'],x=at_top_country['athlet_type'], palette = clrs)
plt.show()
```

# 3.5 Top 20 Country

```python
counts = athlet["NOC"].value_counts().sort_values(ascending = False).reset_index()
counts.columns = ['country','count']
atlet_top_country=counts.head(20)
#clrs = ['#FF4500' if (x < max(atlet_top_country['count'])) else 'black' for x in atlet_top_country['count'] ]
plt.figure(figsize = (10,8))
sns.barplot(x =atlet_top_country['count'],y=atlet_top_country['country'], palette = "vlag")
```

# 4.2 Initail Idea about Medals Data

```python
medal.info()
print('-'*100)
medal.head()
```

# 4.3 Pie Chart

```python
g = medal["Gold"].sum()
s = medal["Silver"].sum()
b = medal['Bronze'].sum()
d =pd.DataFrame({'total': [g,s,b]},
     index=['gold', 'silver','bronze'])

def label_function(v):
    return f'{v:.0f}%'
print(d)
     
d.plot.pie(y='total', figsize=(5, 5),colors=['#FFD700','#C0C0C0','#CD7F32'],autopct=label_function,textprops={'fontsize': 20})
```

# 3.4 Stacked Bar Graph

```python
data = [medal["Team/NOC"],medal["Gold"], medal["Silver"],medal['Bronze']]

headers = ["Team/NOC","Gold", "Silver","Bronze"]

df3 = pd.concat(data, axis=1, keys=headers)

c=df3.head(15)
fig, ax = plt.subplots(figsize=(12,10))
c.plot(
  x = 'Team/NOC', 
  kind = 'barh', 
  stacked = True, 
  title = 'Stacked Bar Graph', 
  mark_right = True,ax=ax
)
total = c["Gold"] + c["Silver"] + c["Bronze"]
rel = c[c.columns[1:]].div(total, 0)*100 
for n in rel:
    for i, (s, b, p) in enumerate(zip(c.iloc[:, 1:].cumsum(1)[n], 
                                         c[n], rel[n])):
        plt.text(s - b / 2, i, str(np.round(p, 1)) + '%', 
                 va = 'center', ha = 'center',fontsize = 10)
```
